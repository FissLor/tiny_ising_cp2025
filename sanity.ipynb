```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('results_sanity_atom/gcc-O3_8192/table_gcc-O3_8192')
print(df)

# # Load data into a pandas dataframe
# df = pd.read_csv('../results/LAB1/sim_results_modified_code_flags_xorshift.csv')
df['version'] = 'original'
df1 = pd.read_csv('results_sanity_atom/both_8192/table_both_8192')
df1['version'] = 'xoshiro + zero padding'

df2 = pd.read_csv('results_sanity_atom/xoroshift_8192/table_xoroshift_8192')
df2['version'] = 'xoshiro'

df = pd.concat([df, df2, df1])
# Use seaborn to create a line plot
sns.lineplot(x='# Temp', y='E', hue='version' , data=df)

# Set the plot title and axis labels
plt.title('Energy vs Temperature')
plt.xlabel('Temperature')
plt.ylabel('Energy')

# Show the plot
plt.show()
```
