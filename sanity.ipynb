```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('build/vectorizing_rb_update/table')
print(df)

# # Load data into a pandas dataframe
# df = pd.read_csv('../results/LAB1/sim_results_modified_code_flags_xorshift.csv')
df['version'] = 'rb_update'
df1 = pd.read_csv('build/vectorizing_rb_update_boltzmann/table')
df1['version'] = 'rb_update_boltzmann'

df2 = pd.read_csv('build/vectorizing_base/table')
df2['version'] = 'base'


df2 = pd.read_csv('build/vectorizing_simple_update/table')
df2['version'] = 'base'

df3 = pd.read_csv('build/vectorizing_rb_update_boltzmann_intrinsics/table')
df3['version'] = 'intrinsics'


df = pd.concat([df2, df, df1, df3])
# Use seaborn to create a line plot
sns.lineplot(x='# Temp', y='E', hue='version' , data=df)

# Set the plot title and axis labels
plt.title('Energy vs Temperature')
plt.xlabel('Temperature')
plt.ylabel('Energy')

# Show the plot
plt.show()
```
